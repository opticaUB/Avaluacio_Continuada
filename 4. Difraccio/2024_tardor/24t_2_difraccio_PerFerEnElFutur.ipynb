```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 50  # mm
y = 10  # mm
s = -50e3 # mm
wavelen = 500e-6 # [mm]
```

```python
sp = 1/(1/s + 1/f)
yp = sp / s * y

print(f"{yp=:.2f} mm")
```

```python
N = np.sqrt(2)**list(range(1, 11))
print(f"{N=}")
```

```python
D_diafra = f / N
rho_Airy = 1.22 * wavelen * sp / D_diafra

plt.figure(figsize=(15, 5))
plt.plot(N,rho_Airy)
plt.plot([N[0], N[-1]], [-yp, -yp], ':')
plt.xticks(N);
```

```python
D_quadrat = D_diafra/2 * np.sqrt(np.pi)
rho_Airy = wavelen * sp / D_quadrat

plt.figure(figsize=(15, 5))
plt.plot(N,rho_Airy)
plt.plot([N[0], N[-1]], [-yp, -yp], ':')
plt.xticks(N);
```
