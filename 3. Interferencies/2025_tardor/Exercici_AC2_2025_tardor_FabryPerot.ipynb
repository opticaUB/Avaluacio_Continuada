## **Exercici 3** d'Avaluació Continuada: Interferències (Poder resolutiu Fabry-Pérot)

<br>


<p style="text-align: justify;">
Per observar el doblet del Cadmi (Cd), centrat en $\lambda=643.8\,\textrm{nm}$ i amb una separació de $\Delta\lambda=0.02\,\textrm{nm}$, es disposa d’un interferòmetre de Fabry-Pérot amb una separació entre els miralls d’$1\,\textrm{mm}$.
</p>

<p style="text-align: justify;">
  <b>a</b>)	Per una reflectivitat de $r=0.5$, representeu gràficament la intensitat transmesa per totes dues longituds d'ona en funció de l'angle d'incidència. També mostreu la suma de les dues intensitats. Comenteu la gràfica i expliqueu si el doblet és resoluble amb aquesta reflectivitat.
</p>

<p style="text-align: justify;">
  <b>b</b>)	Busqueu, per tempteig, la reflectivitat en què la suma de les intensitats estigui en el límit de resolució, segons el criteri de la mitja alçada. Compareu el resultat amb el poder de resolució nominal?
</p>

<p style="text-align: justify;">
  <b>c</b>)	Busqueu, per tempteig, una reflectivitat en què la suma de les intensitats tingui una vall al voltant del 50% dels pics. En aquest cas, serà clarament diferenciable el doblet?
</p>

<p style="text-align: justify;">
  <b>d</b>) Busqueu, per tempteig, una reflectivitat en què la suma de les intensitats mostri un sol pic entorn de transmissió al voltant del doblet. Comenteu el resultat.
</p>

<p style="text-align: justify;">
  <b>e</b>)	Tornant a la reflectivitat mínima nominal (apartat b), per quin angle d'incidència el doblet deixa de ser resoluble del tot (la suma d'intensitats mostra un sol pic)? Mostreu la gràfica al voltant d'aquest angle d'incidència.
</p>
<br>

<br>
