```python
import numpy as np
import matplotlib.pyplot as plt
```

##### INTRODUCCIÓ

Aquest treball és molt obert i buscava que juguéssiu amb les expressions per veure les tendències i treballar una mica la intuïció.

En aquesta proposta de solució pretenc estudiar el camp de visió i el vinyateig (els dos aspectes claus de la il·luminació) en funció de tres paràmetres, mentre que mantinc la resta constants.

 * **Radi de l'ocular**: És obvi que com més ampli sigui l'ocular, més acceptació tindrà de raigs amb angles grans. Per tant, prenc el radi de l'ocular com a variable independent, la que variarà més i sobre la que farem totes les gràfiques.
 * **Radi de l'objectiu**: També juga un paper fonamental amb la il·luminació, ja que com més gran, més raigs entren. Però en voler estudiar el camp i el vinyateig, aquesta variable serà essencialment un factor d'escala. Per aquest motiu mantinc constant l'obertura relativa $N=\frac{f'_{obj}}{r_{obj}}$.
 * **Augment total del microscopi**: En ser un paràmetre tan fonamental, el mantinc constant.
 * **Augment de l'objectiu** i **longitud del tub**: Són dues variables molt lligades, per tant, pot ser interessant com afecten en el camp i la il·luminació, per tant, el prenc com a paràmetres lliures, on estudiarem tres casos de cada per veure les tendències.
 
 Cal dir que aquesta tria és totalment arbitrària i es podria plantejar l'estudi d'una manera totalment diferent.

```python
# Paràmetres del microscopi
#   Quan es tenen moltes variables interdependents i 
#   es vol estudiar el seu comportament, és útil fixar-ne algunes (una o dues),
#   i estudiar com varien les altres. Les que varien ho poden fer de manera 
#   'continua' (amb molts punts) o 'discreta' (amb pocs punts).

# Constants (Fixem algun parametre per fer l'estudi per un cas donat)
Gamma_T = 100  # En aquest cas estudiem el problema per a un augment total donat
N_obj = 1.4    # Fixem també l'obertura numèrica de l'objectiu

# Casos d'estudi (Estudiem diferents casos per veure les tendencies)
beta_obj = np.array([5, 10, 25])      # Augment de l'objectiu
tube_len = np.array([200, 400, 800])  # [mm] Longitud del tub

# Variables (Busquem el paràmetre més rellevant en el resultat final)
r_ocu = np.linspace(1, 50)  # [mm] Rang pel radi de l'ocular

# Paràmetres derivats de beta
Gamma_ocu = Gamma_T / beta_obj
f_ocu = 250 / Gamma_ocu      # [mm] Focal de l'ocular
f_obj = tube_len / beta_obj  # [mm] Focal de l'objectiu
r_obj = f_obj / N_obj        # [mm] radi de l'objectiu
```

```python
# Càlculs bàsics

y = lambda h, b: 2 * h / b  # Mida en l'espai objecte (y = 2h/beta) 

# alçada per la que passa el raig que marca la il·luminació plena sobre el pla focal objecte de l'ocular
hp = lambda roc, rob, foc, s_prime, total_len: (roc * s_prime - rob * foc) / total_len

# alçada per la que passa el raig que marca la il·luminació límit sobre el pla focal objecte de l'ocular
hl = lambda roc, rob, foc, s_prime, total_len: (roc * s_prime + rob * foc) / total_len
```

```python
# Gràfics
fig, axs = plt.subplots(3, 3, figsize=(10,10))  #, sharex=True, sharey=True)

for i, beta in enumerate(beta_obj):
    for j, t in enumerate(tube_len):
        ax = axs[i][j]
        ax.set_title(fr"$\beta'={beta}$ ; $t={t/10:.0f}$ cm ; $\Gamma_o={Gamma_ocu[i]:.0f}$", x=0.5, y=0.85, backgroundcolor='white')
        x_pos = 'bottom' if i == 2 else ('top' if i == 0 else None)
        if x_pos:
            ax.xaxis.set_label_position(x_pos)
            ax.xaxis.set_ticks_position(x_pos)
            ax.set_xlabel(r'$r_{ocu}$ [mm]')
        else:
            ax.xaxis.set_ticks([])
        y_pos = 'left' if j == 0 else ('right' if j == 2 else None)
        if y_pos:
            ax.yaxis.set_label_position(y_pos)
            ax.yaxis.set_ticks_position(y_pos)
            ax.set_ylabel('Sample size $2y$ [mm]')
        else:
            ax.yaxis.set_ticks([])

        s_imag = f_obj[i] + t
        L = f_obj[i] + t + f_ocu[i]

        ax.set_xlim(r_ocu.min(), r_ocu.max())
        ax.set_ylim(0, 10)
        ax.plot(r_ocu, 2*y(hp(r_ocu, r_obj[i], f_ocu[i], s_imag, L), beta), label=r'$y_P$')
        ax.plot(r_ocu, 2*y(r_ocu * s_imag / L, beta), label=r'$y_M$')
        ax.plot(r_ocu, 2*y(hl(r_ocu, r_obj[i], f_obj[i], s_imag, L), beta), label=r'$y_L$')
        ax.legend(loc='lower right')  # bbox_to_anchor=(0.4, 0.45))
        
plt.tight_layout(h_pad=1.2, w_pad=0.8)
plt.show()
```

El que es pot veure és que en augmentar la focal de la lent de tub, tot mantenint l'augment de l'objectiu, el vinyeteig decreix: ja que el diametre plenament il·luminat ($2y_P$) és més semblant al diametre límit ($2y_L$).

Per altra banda, en incrementar l'augment de l'objectiu, el camp de visió es redueix força o, dit d'una altra maenra, cal més esforç en augmentar la mida de l'ocular per aconseguir el mateix camp de visió (pendent de les gràfiques).
